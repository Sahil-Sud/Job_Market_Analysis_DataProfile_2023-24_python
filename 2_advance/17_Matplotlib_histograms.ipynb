```python
import pandas as pd
import matplotlib.pyplot as plt
import ast 
import numpy as np
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_skills']  = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x) 

```

```python
df_US_DA = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == 'United States')].copy()
df_US_DA
```

```python
df_US_DA['salary_year_avg'].plot(kind = 'hist', bins = 40, edgecolor = 'black')
plt.xlim(0,250000)

ax= plt.gca()

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x , pos: f'${int(x/1000)}k'))
plt.title('Distribution of salary for Data Analyst role across U.S.')
plt.xlabel('Yearly Salary')
plt.ylabel('Number of jobs')
```
