```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
pt = df.pivot_table(index = 'job_country', columns='job_title_short', values='salary_year_avg' , aggfunc= 'median' )
```

```python
jt = ['Data Analyst', 'Data Engineer','Data Scientist']
```

```python
c = df['job_country'].value_counts().sort_values(ascending=False).head(6).index
```

```python
f_pt = pt.loc[c,jt]
f_pt
```

```python
f_pt.plot(kind = 'bar')
plt.xlabel('')
plt.ylabel('median salary')
plt.xticks(rotation = 45, ha='right')
plt.title('Median Salary by top 6 Country and top 3 Job Title')
plt.show()
```
