```python
import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y)
plt.show()
```

```python
df.job_posted_date.dt.date.value_counts().plot(kind='bar')
plt.show()
```

```python
date_counts = df.job_posted_date.value_counts()
```

```python
type(date_counts)
```

```python
date_counts
date_counts = date_counts.sort_index()
date_counts
```

```python
plt.plot(date_counts.index,date_counts.values)
plt.show()
```

```python
df['job_posted_month']= df.job_posted_date.dt.month
df.head(10)
```

```python
monthly_count = df.job_posted_month.value_counts()
monthly_count = monthly_count.sort_index()
```

```python
plt.plot(monthly_count.index,monthly_count.values)
plt.show()
```

```python
job_count = df.job_title_short.value_counts().sort_values(ascending = False)
job_count
```

```python
plt.bar(job_count.index, job_count.values)
plt.title('Posting by job title')
plt.legend()
plt.xlabel('Job title')
plt.ylabel('Number of postings')
plt.xticks(rotation = 45 ,ha = 'right')
plt.show()
```

```python
job_count = job_count.sort_values(ascending = True)
plt.barh(job_count.index, job_count.values)
plt.show()
```

```python
job_count.plot(kind='pie')
plt.show()
```
